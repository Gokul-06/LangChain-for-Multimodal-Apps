```python
source_url = 'https://tile.loc.gov/storage-services/service/pnp/fsa/8b29000/8b29500/8b29516v.jpg'
destination_dir = 'kaggle'  # The directory where you want to save the image
image_filename = os.path.basename(source_url)  # This will extract '8b29516v.jpg' as the filename

# Combine the directory and filename to create the destination path
destination_path = os.path.join(destination_dir, image_filename)
```

```python
# Ensure the destination directory exists, create if it doesn't
os.makedirs(destination_dir, exist_ok=True)

# Fetch and save the image
response = requests.get(source_url)
if response.status_code == 200:
    with open(destination_path, 'wb') as f:
        f.write(response.content)
    print(f"Image successfully downloaded to {destination_path}")
else:
    print(f"Failed to download the image. Status code: {response.status_code}")
```

```python
import os
from langchain_community.document_loaders import ImageCaptionLoader

# Set the path to the 'kaggle' directory
path = 'kaggle'

# Ensure the path exists
if not os.path.exists(path):
    print(f"The specified path '{path}' does not exist.")
else:
    # Get image URIs with .jpg extension only
    image_uris = sorted(
        [
            os.path.join(path, image_name)
            for image_name in os.listdir(path)
            if image_name.endswith(".jpg")
        ]
    )

    # Initialize the ImageCaptionLoader with the image URIs
    loader = ImageCaptionLoader(images=image_uris)

    # Load the documents (images)
    list_docs = loader.load()
```
